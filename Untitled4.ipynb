```python
plt.grid()
plt.xlabel( 'Daily' )
plt.ylabel( 'Electric price' )
plt.plot(df_daily.index, df_daily.ElecPrice, "--", marker = "o")
```

```python
from statsmodels.tsa.stattools import adfuller

```

```python
print("Results of  Dickey-Fuller Test:")
dftest = adfuller(df["ElecPrice"], autolag = "AIC")
# AIC = AKaike information criterion, gives the exact lag value
#If “AIC” (default) or “BIC”, then the number of lags is chosen to minimize
# the corresponding information criterion.
dfoutput = pd.Series(dftest[0:4], index = ["Test Stastistic", "p-value", "#lags used", "Number of Oberservaton Used"])

for key,value in dftest[4].items():
  dfoutput["Critical Value (%s) " %key] = value

print(dfoutput)
```
